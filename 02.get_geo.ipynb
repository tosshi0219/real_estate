```python
# coding=utf-8
import pandas as pd
import time
import pickle
import requests
import xmltodict
```

```python
spec=pd.read_csv('spec.csv')
address=spec['所在地']
```

```python
def latlng(ad):
    url = 'http://www.geocoding.jp/api/'
    payload = {'q': ad}
    result = requests.get(url, params=payload)
    resultdict = xmltodict.parse(result.text)
    return resultdict["result"]["coordinate"]["lat"],resultdict["result"]["coordinate"]["lng"]
```

```python
lats=[]
lngs=[]
for i in range(3413,5000):
    lat,lng=latlng(address[i])
    print(i,lat,lng)
    lats.append(lat)
    lngs.append(lng)
    time.sleep(5)
```

```python
# 取得したページのhtmlをpickle化
with open('lats1360-3413.pickle','wb') as f:
    pickle.dump(lats, f)
```

```python
# 取得したページのhtmlをpickle化
with open('lngs1360-3413.pickle','wb') as f:
    pickle.dump(lngs, f)
```
